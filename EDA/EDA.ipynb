# EDA on the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url="https://raw.githubusercontent.com/KelvinYQC/msia420PA_project/6bb5fe63050105194819f99c90476486076e5227/Data/Hotel%20Reservations.csv"
dat=pd.read_csv(url)
print(dat.shape)
dat.head()
```

```python
dat.market_segment_type.value_counts()
```

## Checking missing values

```python
(dat.isnull().sum()/len(dat))
```

There is no missing value in this dataset

```python
cov = dat.corr().round(2)
plt.figure(figsize = (15,10))
sns.heatmap(cov, annot = True,cmap="RdBu_r", center = 0)
```

There are no features highly correlated with each other.

## Check the distribution for the response variable.

```python
dat['booking_status'].hist()
```

```python
dat.hist(figsize=(16,16))
plt.show()
```

## Numerical features and booking status

```python
plt.figure(figsize = (25, 20))
plt.subplot(2,2,1)
sns.kdeplot(x='lead_time', hue='booking_status', shade=True, data=dat)

plt.subplot(2,2,2)
sns.kdeplot(x = 'avg_price_per_room', hue = 'booking_status', shade=True, data = dat)

plt.subplot(2,2,3)
sns.kdeplot(x='arrival_month', hue='booking_status', shade=True, data=dat)

plt.subplot(2,2,4)
sns.kdeplot(x = 'arrival_date', hue = 'booking_status', shade=True, data = dat)
```

## Categorical featuers and booking status

```python
plt.figure(figsize = (15, 10))
plt.subplot(2,2,1)
sns.countplot(x='repeated_guest', hue='booking_status', data=dat)

plt.subplot(2,2,2)
sns.countplot(x='required_car_parking_space', hue='booking_status', data=dat)

plt.subplot(2,2,3)
sns.countplot(x='room_type_reserved', hue='booking_status', data=dat)

plt.subplot(2,2,4)
sns.countplot(x='market_segment_type', hue='booking_status', data=dat)


```
