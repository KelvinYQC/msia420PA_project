# EDA on the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dat = pd.read_csv("/Users/kelvin/Desktop/Winter2023/MSiA_420_pa/project/Hotel Reservations.csv")
print(dat.shape)
dat.head()
```

## Checking missing values

```python
(dat.isnull().sum()/len(dat))
```

There is no missing value in this dataset

```python
cov = dat.corr().round(2)
plt.figure(figsize = (15,10))
sns.heatmap(cov, annot = True,cmap="RdBu_r", center = 0)
```

There are no features highly correlated with each other.

## Check the distribution for the response variable.

```python
dat['booking_status'].hist()
```

```python
dat.hist(figsize=(16,16))
plt.show()
```

Numerical features and booking status

```python
plt.figure(figsize = (25, 20))
plt.subplot(2,2,1)
sns.kdeplot(x='lead_time', hue='booking_status', shade=True, data=dat)

plt.subplot(2,2,2)
sns.kdeplot(x = 'avg_price_per_room', hue = 'booking_status', shade=True, data = dat)

plt.subplot(2,2,3)
sns.kdeplot(x='arrival_month', hue='booking_status', shade=True, data=dat)

plt.subplot(2,2,4)
sns.kdeplot(x = 'arrival_date', hue = 'booking_status', shade=True, data = dat)
```
